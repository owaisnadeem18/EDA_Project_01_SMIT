# **2. Which team has the highest win percentage?**

```python
cwc_2019["Winner"] = cwc_2019["Result"].str.extract(r"^(.*?)\s+won")
```

```python
plt.figure(figsize=(24, 8))

sns.countplot(x = "Winner" , data = cwc_2019 )
```

# **6. Which stadium hosted the most matches?**

```python
plt.figure(figsize=(34,12))

stadium_counts = cwc_2019["Ground / Location"].value_counts()
top_stadium = stadium_counts.idxmax()

total_number_of_matches_at_top_stadium = stadium_counts.max()

total_number_of_matches_at_top_stadium

sns.barplot(x = stadium_counts.index , y = stadium_counts.values)

plt.title(f"Most matches were hosted at {top_stadium} which were ({total_number_of_matches_at_top_stadium} matches)", fontsize=14)

plt.xlabel("Ground")
plt.ylabel("Number of Matches")
```

# **7. How many matches ended without a result?**

```python
cwc_2019.head()
```

```python
no_result_matches = cwc_2019[cwc_2019["Winner"].isna()]

# Ek naya column bana dete hain sirf "No Result" label ke liye
no_result_matches["Result_Type"] = "No Result"

sns.countplot(x="Result_Type", data=no_result_matches)

plt.title("📊 Matches ended without results")
plt.xlabel("Result Type")
plt.ylabel("Number of Matches")
plt.show()
```
